# 나만의 코퍼스 만들기
- txt 데이터 수집
- txt 인코딩 utf-8으로 저장
- txt 파일명에 날짜, 이름 등 정보 부여
- txt 드라이브 업로드
- 폴더명 읽기
- for문 활용 파일명 읽고, 날짜, 이름, 텍스트 리스트 생성
- 딕셔너리와 zip()활용하여 코퍼스 완성
  * 주의사항: key는 영문. 한글은 불가 함.

```python
import typing_extensions
# lyrics

import os

dir_path = '/content/drive/MyDrive/Colab Notebooks/2022_2_textdataanalysis/corpus_rawdata/lyrics'

titles = []
names = []
dates = []
texts = []

for filename in os.listdir(dir_path):
  print(filename)
  tmp1 = filename.replace('.txt','')  # 확장자명 제거
  tmp2 = tmp1.split('.')  # 피리오드(.)기준으로 분리하여 문자 리스트 반환
  # print(tmp2)
  title, name, date = tmp2[0], tmp2[-2], tmp2[-1]

  titles.append(title)
  names.append(name)
  dates.append(date)
  
  with open(dir_path + '/' + filename, 'r') as f:
    text = f.readlines()[3:]
    text_all = ' '.join(text)
    # print(filename)
    # print(text_all)
    texts.append(text_all)

# print(len(names), len(dates), len(titles), len(texts)) # 동일한 길이
# print(names)
# print(dates)
# print(titles)

lyrics_corpus_dict = {}
for name, date, title, text in zip(names, dates, titles, texts):
  #print(name, date, title)
  lyrics_corpus_dict[date] = [name, title, text_all]   # dict key 한글 불가

print(lyrics_corpus_dict)
print(lyrics_corpus_dict.keys())
print(list(lyrics_corpus_dict.values())[0])
print(list(lyrics_corpus_dict.values())[1])
print(list(lyrics_corpus_dict.values())[2])
```

```python
# poem
import os

dir_path = '/content/drive/MyDrive/Colab Notebooks/2022_2_textdataanalysis/corpus_rawdata/poem'

titles = []
names = []
texts = []

for filename in os.listdir(dir_path):
  print(filename)
  tmp1 = filename.replace('.txt','')  # 확장자명 제거
  tmp2 = tmp1.split('_')  # 언더바(_)기준으로 분리하여 문자 리스트 반환
  #print(tmp2)
  name, title = tmp2[0], tmp2[1]
  names.append(name)
  titles.append(title)

  with open(dir_path + '/' + filename, 'r', encoding = 'cp949') as f:
    text = f.read()
    #print(text)
    texts.append(text)

#print(len(names), len(titles), len(texts)) # 동일한 길이

poem_corpus_dict = {}
i = 0
for name, title, text in zip(names, titles, list(texts)):
  #print(i, name, title, text)
  poem_corpus_dict[i] = [name, title, text]  # dict key 한글 불가
  i += 1

# print(poem_corpus_dict)
print(poem_corpus_dict.keys())
print(list(poem_corpus_dict.values())[0])
print(list(poem_corpus_dict.values())[1])
print(list(poem_corpus_dict.values())[2])
```

```python
# speech
import os

dir_path = '/content/drive/MyDrive/Colab Notebooks/2022_2_textdataanalysis/corpus_rawdata/speech'

names = []
dates = []
texts = []
for filename in os.listdir(dir_path):
  print(filename)
  tmp1 = filename.replace('.txt','')  # 확장자명 제거
  tmp2 = tmp1.split('_')  # 언더바(_)기준으로 분리하여 문자 리스트 반환
  date, name = tmp2[0], tmp2[1]
  dates.append(date)
  names.append(name)
  
  with open(dir_path + '/' + filename, 'r', encoding='utf8') as f:
    text = f.read()
    texts.append(text)
    #print(text)

speech_corpus_dict = {}
for name, date, text in zip(names, dates, texts):
  speech_corpus_dict[name] = [date, text]

print(speech_corpus_dict)
print(speech_corpus_dict.keys())
print(list(speech_corpus_dict.values())[0])
print(list(speech_corpus_dict.values())[1])
print(list(speech_corpus_dict.values())[2])
```
